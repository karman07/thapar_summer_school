```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression, RFE
from sklearn.ensemble import GradientBoostingRegressor, StackingRegressor
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.base import clone
from sklearn.pipeline import Pipeline
import optuna
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("train.csv")
df = df.drop(['Row#'], axis=1)

X = df.drop("output", axis=1)
y = df["output"]
```

```python
def preprocess_data(X, y):

    Q1 = X.quantile(0.25)
    Q3 = X.quantile(0.75)
    IQR = Q3 - Q1
    mask = ~((X < (Q1 - 1.5 * IQR)) | (X > (Q3 + 1.5 * IQR))).any(axis=1)
    X, y = X[mask], y[mask]

    scaler = RobustScaler()
    X_scaled = scaler.fit_transform(X)

    selector = SelectKBest(score_func=f_regression, k='all')
    X_selected = selector.fit_transform(X_scaled, y)

    pca = PCA(n_components=0.95)
    X_pca = pca.fit_transform(X_selected)

    return X_pca, y
```

```python
X_processed, y_processed = preprocess_data(X, y)
```

```python
gbr = GradientBoostingRegressor()
xgb = XGBRegressor(verbosity=0)
lgb = LGBMRegressor()
cat = CatBoostRegressor(verbose=0)

models = {
    "GradientBoosting": gbr,
    "XGBoost": xgb,
    "LightGBM": lgb,
    "CatBoost": cat
}
```

```python
def evaluate_models(X, y, models):
    print("Model Evaluation Before Tuning (Cross-validated RMSE):")
    for name, model in models.items():
        scores = cross_val_score(model, X, y, scoring="neg_root_mean_squared_error", cv=5)
        print(f"{name}: {-np.mean(scores):.4f} ± {np.std(scores):.4f}")

evaluate_models(X_processed, y_processed, models)
```

```python
def optimize_model(name, model_type, X, y):
    def objective(trial):
        if model_type == "gbr":
            model = GradientBoostingRegressor(
                n_estimators=trial.suggest_int("n_estimators", 50, 300),
                learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
                max_depth=trial.suggest_int("max_depth", 3, 10)
            )
        elif model_type == "xgb":
            model = XGBRegressor(
                n_estimators=trial.suggest_int("n_estimators", 50, 300),
                learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
                max_depth=trial.suggest_int("max_depth", 3, 10),
                subsample=trial.suggest_float("subsample", 0.5, 1.0),
                verbosity=0
            )
        elif model_type == "lgb":
            model = LGBMRegressor(
                n_estimators=trial.suggest_int("n_estimators", 50, 300),
                learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
                max_depth=trial.suggest_int("max_depth", 3, 10),
                num_leaves=trial.suggest_int("num_leaves", 20, 100)
            )
        elif model_type == "cat":
            model = CatBoostRegressor(
                iterations=trial.suggest_int("iterations", 100, 300),
                learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
                depth=trial.suggest_int("depth", 3, 10),
                verbose=0
            )

        score = cross_val_score(model, X, y, scoring="neg_root_mean_squared_error", cv=3).mean()
        return -score

    study = optuna.create_study(direction="minimize")
    study.optimize(objective, n_trials=30)
    print(f"\n{name} Best Score: {study.best_value:.4f}")
    return study.best_trial
```

```python
best_trials = {}

best_trials["GradientBoosting"] = optimize_model("GradientBoosting", "gbr", X_processed, y_processed)
best_trials["XGBoost"] = optimize_model("XGBoost", "xgb", X_processed, y_processed)
best_trials["LightGBM"] = optimize_model("LightGBM", "lgb", X_processed, y_processed)
best_trials["CatBoost"] = optimize_model("CatBoost", "cat", X_processed, y_processed)
```

```python
def build_tuned_models(best_trials):
    tuned_models = {}

    tuned_models["GradientBoosting"] = GradientBoostingRegressor(**best_trials["GradientBoosting"].params)
    tuned_models["XGBoost"] = XGBRegressor(**best_trials["XGBoost"].params, verbosity=0)
    tuned_models["LightGBM"] = LGBMRegressor(**best_trials["LightGBM"].params)
    tuned_models["CatBoost"] = CatBoostRegressor(**best_trials["CatBoost"].params, verbose=0)

    return tuned_models

tuned_models = build_tuned_models(best_trials)
```

```python
evaluate_models(X_processed, y_processed, tuned_models)
```

```python
stack = StackingRegressor(
    estimators=[
        ("gbr", tuned_models["GradientBoosting"]),
        ("xgb", tuned_models["XGBoost"]),
        ("lgb", tuned_models["LightGBM"]),
        ("cat", tuned_models["CatBoost"])
    ],
    final_estimator=Ridge()
)

stack_score = cross_val_score(stack, X_processed, y_processed, scoring="neg_root_mean_squared_error", cv=5)
print(f"\nStacked Regressor RMSE: {-np.mean(stack_score):.4f} ± {np.std(stack_score):.4f}")
```

```python
r2_scores = cross_val_score(stack, X_processed, y_processed, scoring="r2", cv=5)
```

```python
print(f"  R²  :  {np.mean(r2_scores):.4f} ± {np.std(r2_scores):.4f}")
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.decomposition import PCA

df = pd.read_csv("train.csv")

X_raw = df.drop(["Row#", "output"], axis=1)
y_raw = df["output"]

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X_raw)

selector = SelectKBest(score_func=f_regression, k='all')
X_selected = selector.fit_transform(X_scaled, y_raw)

pca = PCA(n_components=0.95)
X_final = pca.fit_transform(X_selected)

stack.fit(X_final, y_raw)

test_df = pd.read_csv("test.csv")
test_ids = test_df["id"]

X_test_raw = test_df.drop(["id", "Row#"], axis=1)
X_test_scaled = scaler.transform(X_test_raw)
X_test_selected = selector.transform(X_test_scaled)
X_test_final = pca.transform(X_test_selected)

predictions = stack.predict(X_test_final)

submission_df = pd.DataFrame({
    "id": test_ids,
    "prediction": predictions
})

submission_df.to_csv("submission.csv", index=False)
```
